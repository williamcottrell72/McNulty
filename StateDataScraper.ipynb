```python
states='http://www.50states.com/abbreviations.htm'
response = requests.get(states)
soup2=BeautifulSoup(response.text,'html.parser')
```

```python
tbl=soup2.find_all('table',class_="spaced stripedRows")[0].find_all('tr')
```

```python
state_dct={}
for item in tbl[1:]:
    try:
        state_dct[item.find_all('td')[0].text]=item.find_all('td')[1].text
    except(IndexError):
        pass
```

```python
def abbreviate(x):
    return state_dct[x]
```

```python
table=soup.find_all('div',class_="gov-data-table")
```

```python
tr_list=[]
for item in table:
    trs=item.find_all('tr',role="row")
    tr_list.append(trs)
```

```python
tr_clean_list=tr_list[0][2:]
#soup2=BeautifulSoup(tr_clean_list[2],'html.parser')
```

```python
[x.text for x in tr_clean_list[2].find_all('td')]
```

```python
dct={'location':[],'density':[],'pop':[],'area':[]}
for item in tr_clean_list:
    city_state=item.find_all('td')[0].text
    city=city_state.split(',')[0]
    state_long=city_state.split(',')[1].strip()
    state=abbreviate(state_long)
    loc=city+', '+state
    dct['location'].append(loc)
    #dct['city'].append(city)
    #dct['state'].append(state)
    dct['density'].append(int(item.find_all('td')[1].text.replace(',','')))
    dct['pop'].append(int(item.find_all('td')[2].text.replace(',','')))
```

```python
df_cities=pd.DataFrame(dct,columns=['location','density','pop'])
```

```python
with open('city_density.pkl','wb') as picklefile:
    pickle.dump(df_cities,picklefile)
```
